### Change date columns to datetime.


```python
def date_totime(time_str):
    time_dt = dt.datetime.strptime(time_str, "%d/%m/%Y")
    return time_dt
```

```python
for column in ['date_scrapped','published','expire']:
    data[column] = data[column].apply(date_totime)
```
